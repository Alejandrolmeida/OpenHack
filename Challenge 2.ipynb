## Padding & Resize

```python
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image, ImageOps

# Specify Image
desired_size = 128
im_pth = "gear_images/insulated_jackets/866332.jpeg"

# Load Image
im = Image.open(im_pth)
old_size = im.size  # old_size[0] is in (width, height) format

# Calculate new size
ratio = float(desired_size)/max(old_size)
new_size = tuple([int(x*ratio) for x in old_size])

# Resize
im = im.resize(new_size, Image.ANTIALIAS)

# Compose new image with padding
new_im = Image.new("RGB", (desired_size, desired_size), 0xFFFFFF)
new_im.paste(im, ((desired_size-new_size[0])//2,
                    (desired_size-new_size[1])//2))
```

```python
plt.imshow(new_im)
plt.show()
```

## Equalizing

```python
import numpy as np

img_pil = Image.fromarray(np.uint8(new_im))
img_pil = ImageOps.equalize(img_pil)
plt.imshow(img_pil)
```

```python
data = np.asarray(im).ravel()
fig = plt.figure(figsize=(12,6))
plt.subplot(121)
plt.title("Original")
plt.hist(data)
plt.xlim(xmin=0, xmax = 255.0)

plt.subplot(122)
plt.title("Equalized")
data = np.asarray(img_pil).ravel()
plt.hist(data)
plt.xlim(xmin=0, xmax = 255.0)

plt.show()
```
